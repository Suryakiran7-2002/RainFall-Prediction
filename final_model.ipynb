```python
import pandas as pd
```

```python
data  = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
dummpies = pd.get_dummies(data.SUBDIVISION)
dummpies
```

```python
merged = pd.concat([data,dummpies],axis = 'columns')
merged
```

```python
final = merged.drop(["SUBDIVISION"], axis = 'columns')
final
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = final.drop(['y'],axis =  'columns')
y = final.iloc[:,3]

m12 = X.iloc[:,0]
m12 = m12 ** 2

m22 = X.iloc[:,1]
m22 = m22 ** 2

m32 = X.iloc[:,2]
m32 = m32 ** 2

mm = X.iloc[:,0] * X.iloc[:,1] * X.iloc[:,2]

X = pd.concat([X,m12],axis = 'columns')
X = pd.concat([X,m22],axis = 'columns')
X = pd.concat([X,m32],axis = 'columns')
X = pd.concat([X,mm],axis = 'columns')

```

```python
avg = (X.iloc[:,0] + X.iloc[:,1] + X.iloc[:,2])  ** 2
X = pd.concat([X,avg],axis = 'columns')

ind_2 = (X.iloc[:,0] + X.iloc[:,1] ) ** 2 +  (X.iloc[:,1] + X.iloc[:,2] ) ** 2 +  (X.iloc[:,2] + X.iloc[:,0] ) ** 2
X = pd.concat([X,avg],axis = 'columns')
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
